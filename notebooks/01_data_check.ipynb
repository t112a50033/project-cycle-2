1. Is the proportion of students who feel sad or hopeless different from 0.30?

### 2-1 SadOrHopeless
- Variable name: SadOrHopeless
- Meaning: Whether the student felt sad or hopeless
- Valid codes:
  - 1 = Yes
  - 2 = No
- Success definition:
  - success = code 1
  - failure = code 2
- Benchmark for one-sample proportion test:
  - p0 = 0.30

### 2-2 BMIPCT
- Variable name: BMIPCT
- Meaning: BMI percentile
- Benchmark for one-sample mean test:
  - μ0 = 65

## 7. Missing Value Check

```python
df["SadOrHopeless"].isna().sum()
df["BMIPCT"].isna().sum()
```

Missing values are identified and will be handled in the data cleaning step.

## 8. Data Cleaning and Recoding: SadOrHopeless

```python
sad_df = df[["SadOrHopeless"]].copy()   
sad_df = sad_df[sad_df["SadOrHopeless"].isin([1, 2])]
sad_df["Sad_binary"] = sad_df["SadOrHopeless"].map({1: 1, 2: 0})

sad_df[["SadOrHopeless", "Sad_binary"]].head()
```

Only valid responses (codes 1 and 2) are retained for SadOrHopeless. The variable is recoded into a binary format, where 1 represents "Yes" (feeling sad or hopeless) and 0 represents "No". This binary variable will be used in the proportion analysis.

## 9. Data Cleaning: BMIPCT

```python
bmi_df = df[["BMIPCT"]].dropna().copy()
bmi_df.head()
```

Missing values in BMIPCT are removed to ensure that the mean analysis is based only on valid observations. The cleaned dataset will be used for calculating the sample mean and conducting inference.

## 10. Joint Dataset
### 10-1 Prepare Dataset for Joint Analysis

```python
joint_df = df[["SadOrHopeless", "BMIPCT"]].copy()
joint_df = joint_df[joint_df["SadOrHopeless"].isin([1, 2])]
joint_df = joint_df.dropna(subset=["BMIPCT"])
joint_df["Sad_binary"] = joint_df["SadOrHopeless"].map({1: 1, 2: 0})

joint_df.head()
```

A combined dataset is created by keeping only valid SadOrHopeless responses and non-missing BMIPCT values. This dataset will be used for additional EDA to explore the relationship between BMI percentile and the likelihood of feeling sad or hopeless.

## 11. Sample Sizes

```python
print("SadOrHopeless sample size:", len(sad_df))
print("BMIPCT sample size:", len(bmi_df))
print("Joint sample size for extra EDA:", len(joint_df))
```

After cleaning the data, we obtained the final sample sizes for both the proportion and mean analyses. These values will be used in subsequent EDA and inference steps.

## 12. Save Processed Datasets

```python
sad_df.to_csv("../data/processed/sad_cleaned.csv", index=False)
bmi_df.to_csv("../data/processed/bmi_cleaned.csv", index=False)
joint_df.to_csv("../data/processed/joint_cleaned.csv", index=False)
```

The cleaned datasets are saved in the processed folder. This ensures that the data cleaning steps are reproducible and that the analysis can be performed consistently using the processed data.
